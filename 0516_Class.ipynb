```python
from nltk.corpus import gutenberg

# iterate through all the file IDs in the Gutenberg corpus
for fileid in gutenberg.fileids():
    # get the raw text of the file
    text = gutenberg.raw(fileid)

    # calculate the number of characters in the file
    num_chars = len(text)

    # tokenize the text into words
    words = gutenberg.words(fileid)

    # calculate the number of words in the file
    num_words = len(words)

    # tokenize the text into sentences
    sentences = gutenberg.sents(fileid)

    # calculate the number of sentences in the file
    num_sents = len(sentences)

    # print the number of characters, words, sentences, and the file ID
    print(num_chars, num_words, num_sents, fileid)
```

```python
import nltk
from nltk.corpus import inaugural
import matplotlib.pyplot as plt

cfd = nltk.ConditionalFreqDist(
  (target, fileid[:4])
  for fileid in inaugural.fileids()
  for w in inaugural.words(fileid)
  for target in ['america', 'citizen']
  if w.lower() == target )
plt.figure(figsize=(10,5))
cfd.plot()
```
